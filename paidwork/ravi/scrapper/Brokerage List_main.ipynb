```python
url = "https://pplinfo2.dre.ca.gov/PPLInfo/PplInfoStart?LicenseID=02051216" #save the URL in a value called 'url'
response = requests.get(url) #use the "'get' function from the 'requests' library to call in the URL.
response.raise_for_status() #use the raise_for_status function from the 'requests' library to check the status of the url connection.
soup = BeautifulSoup(response.text, "html.parser") #BeautifulSoup is used here to parse HTML and store it in the variable 'soup'.
#print(soup.prettify)
```

```python
brokerage_list = soup.find_all('div', class_='row')
filtered_brokers = []
found_broker_associates = False

for row in brokerage_list:
    title_div = row.find('div', class_='col-lg-3')
    
    if not found_broker_associates and title_div and 'Broker Associates' in title_div.get_text():
        found_broker_associates = True
        
    if found_broker_associates:
        broker_div = row.find('div', class_='col-lg-9')
        if broker_div:
            filtered_brokers.append(broker_div)

print(filtered_brokers)
```

```python


print("| License Number | License Expiration Date | Agent Name |")
print("|---------------|------------------------|------------|")

for div in filtered_brokers:
    lines = [line.strip() for line in div.get_text().split('\n') if line.strip()]
    
    if len(lines) >= 2:
        try:
            license_part, name = lines[0].split(' - ', 1)
            date = lines[1].split(': ')[1]
            
            print(f"| {license_part.strip()} | {date.strip()} | {name.strip()} |")
        except ValueError:
            continue





```

```python
brokerage_data = []

for div in filtered_brokers:
    lines = [line.strip() for line in div.get_text().split('\n') if line.strip()]
    
    if len(lines) >= 2:
        try:
            license_part, name = lines[0].split(' - ', 1)
            date = lines[1].split(': ')[1]
            
            brokerage_data.append([
                license_part.strip(),
                date.strip(), 
                name.strip()
            ])
        except ValueError:
            continue

import csv

with open('brokerage_list.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['License Number', 'License Expiration Date', 'Agent Name'])
    writer.writerows(brokerage_data)

print("Data saved to brokerage_list.csv")
```
